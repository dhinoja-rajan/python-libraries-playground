# Bar Plot

```python
x=["Python", "Numpy","Pandas", "Matplotlib"]
y=[95, 60, 70, 85]
z = [10,20,30,40]
c = ['r', 'g', 'b', 'y']
plt.xlabel("Languages/Libraries", fontsize = 15)
plt.ylabel("Importance", fontsize = 15)
plt.title("Chart of Importance of every Languages/Libraries", fontsize=20, fontweight = 800)
# plt.bar(x,y,color=c, width=0.4)
# plt.bar(x,y,color=c, width=0.4, align="edge", edgecolor="m", linewidth=5)
# plt.bar(x,y,color=c, width=0.4, align="edge", edgecolor="m", linewidth=5, linestyle=":")
plt.bar(x,y,color=c, width=0.4, align="edge", edgecolor="m", linewidth=5, linestyle=":", alpha=0.5)
plt.show()
```

```python
plt.bar(x,y,color=c, width=0.4, align="edge", linewidth=5, label='popularity1')
plt.bar(x,z, width=0.4, align="edge", linewidth=5, label='popularity2')
plt.legend()
plt.show()
```

```python

x=["Python", "Numpy","Pandas", "Matplotlib"]
y=[95, 60, 70, 85]
z = [10,20,30,40]
c = ['r', 'g', 'b', 'y']

p = np.arange(len(x))
width=0.2
p1= [j+width for j in p]

plt.xlabel("Languages/Libraries", fontsize = 15)
plt.ylabel("Importance", fontsize = 15)
plt.title("Chart of Importance of every Languages/Libraries", fontsize=20, fontweight = 800)

plt.bar(p,y,width, color=c, align="edge", linewidth=5, label='popularity1')
plt.bar(p1,z,width,  align="edge", linewidth=5, label='popularity2')

# plt.xticks(p1, x)
plt.xticks(p+width, x, rotation=30)
plt.legend(loc='upper center')
plt.show()
```

```python

x=["Python", "Numpy","Pandas", "Matplotlib"]
y=[95, 60, 70, 85]
z = [10,20,30,40]
c = ['r', 'g', 'b', 'y']

p = np.arange(len(x))
width=0.8
p1= [j+width for j in p]

plt.xlabel("Languages/Libraries", fontsize = 15)
plt.ylabel("Importance", fontsize = 15)
plt.title("Chart of Importance of every Languages/Libraries", fontsize=20, fontweight = 800)

plt.bar(p,y,width, color=c, align="edge", linewidth=5, label='popularity1')
plt.barh(p1,z,width,  align="edge", linewidth=5, label='popularity2')

# plt.xticks(p1, x)
plt.xticks(p+width, x, rotation=30)
plt.legend(loc='upper center')
plt.show()
```

# Scatter Plot

```python
day = [1,2,3,4,5,6,7]
num = [4,6,3,5,1,5,2]
# colors = ['r', 'g', 'b', 'm', 'y','c', 'b' ]
colors = [10, 48, 24, 98, 37, 92, 18]
sizes = [200, 50, 250, 80, 180, 100, 140]

plt.title("Scatter Plot", fontsize=20, fontweight=600)
plt.xlabel("Day", fontsize=15)
plt.ylabel("Numbers", fontsize=15)

# plt.scatter(day, num,c=colors, s=sizes)
# plt.scatter(day, num,c=colors, s=sizes, marker= "*")
# plt.scatter(day, num,c=colors, s=sizes, alpha=0.5, edgecolor ="r", linewidth=3)
plt.scatter(day, num,c=colors, s=sizes, cmap="viridis")
t= plt.colorbar()
t.set_label("colorbar")
plt.show()
```

```python
#Nested Pie Chart {Another way}

x= [10, 20, 30, 40, 50, 60]
x1 = [60, 50, 40, 30, 20, 10]
y = ["ArithmeticError", "AssertionError", "AttributeError", "EOFError", "FloatingPointError", "IndexError"]

plt.pie(x, labels=y, radius=1)
cr = plt.Circle(xy=(0, 0), radius=0.6, facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
