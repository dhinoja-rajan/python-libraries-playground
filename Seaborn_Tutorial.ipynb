# Seaborn Tutorial...

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Seaborn built-in dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# Bar Plot
plt.figure(figsize=(10,4))
sns.barplot(x='sex', y='total_bill', data=df)
plt.title("Bar Plot")
plt.show()
```

```python
# Count Plot
plt.figure(figsize=(10,4))
sns.countplot(x='day', data=df)
plt.title("Count Plot")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10,4))
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Box Plot")
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(10,4))
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', split=True)
plt.title("Violin Plot")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10,4))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex', style='time')
plt.title("Scatter Plot")
plt.show()
```

```python
# Line Plot
plt.figure(figsize=(10,4))
sns.lineplot(x='size', y='total_bill', data=df)
plt.title("Line Plot")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10,4))
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title("Histogram")
plt.show()
```

```python
# KDE Plot
plt.figure(figsize=(10,4))
sns.kdeplot(x=df['total_bill'], hue=df['sex'], fill=True)
plt.title("KDE Plot")
plt.show()
```

```python
# Pair Plot
sns.pairplot(df, hue='sex')
plt.show()
```

```python
# FacetGrid
g = sns.FacetGrid(df, col="sex", row="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
# Setting Color Palette
sns.set_palette("husl")
plt.figure(figsize=(10,4))
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Box Plot with Color Palette")
plt.show()
```

```python
# Setting Theme
sns.set_style("darkgrid")
plt.figure(figsize=(6,4))
sns.lineplot(x='size', y='total_bill', data=df)
plt.title("Line Plot with Theme")
plt.show()
```

```python
# Customizing Matplotlib with Seaborn
plt.figure(figsize=(10,6))
sns.barplot(x='day', y='total_bill', data=df)
plt.title("Total Bill by Day", fontsize=15)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Total Bill", fontsize=12)
plt.show()
```
